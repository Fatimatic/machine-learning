```python
#  creating x_train and y_train variables
x_train  = np.array([1.0,2.0])
y_train=np.array([20,500])
print(f'x_train = {x_train}')
print(f'y_train = {y_train}')
```

```python
m=len(x_train) # x_train.shape[0] also gives length of array
print(f'm which is number of training examples is {m}')

print(f"The number of entries in the given dimesnsion is : {x_train.shape}")
```

```python
i=0
x_i= x_train[i]
y_i= y_train[i]

print(f'(x^({i}),y^({i})) = ({x_i},{y_i})')
```

```python
# plotting the two points
plt.scatter(x_train,y_train,marker='x',c='r')
plt.title('Housing Prices in Geneva')
plt.ylabel('price(in thousands of dollars)')
plt.xlabel('size of houses (1000 sqft)')
plt.show()
```

```python
# As, f(x) = wx+b
w=480; b=-460

def compute_model_output(w,x,b):
    f_wb = np.zeros(m)
    for i in range (m):
        f_wb[i]= w* x[i]+b
    return f_wb

```

```python
temp_f_wb=compute_model_output(w,x_train,b)

plt.plot(x_train, temp_f_wb, c='b',label='Our Prediction')
plt.scatter(x_train,y_train,c='r',marker='x',label='Actual result')
plt.title('Housing Prices in Geneva')
plt.ylabel('price(in thousands of dollars)')
plt.xlabel('size of houses (1000 sqft)')
plt.legend()
plt.show()
```
